# contourf 練習

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x,y):
    return(x+1/2+x**5+y**3)*np.exp(-x**2-y**2)
```

```python
n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
```

```python
plt.contourf(X,Y,f(X,Y), 10, alpha=0.75, cmap=plt.cm.cool)
C = plt.contour(X,Y,f(X,Y), 10, colors='black', linewidth=0.5)
plt.clabel(C, inline=True, fontsize=10)
```
